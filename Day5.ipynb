Classes and Objects:

    Design a class Student with attributes like name, roll_number, and marks (list of scores). Include methods to calculate the total marks and average score.
    Create a class Rectangle with attributes for width and height. Implement methods to calculate the area and perimeter.
    Define a class Employee with attributes like name, salary, and department. Include methods to apply a raise and display employee information.


```python
class Student:
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks

    def total_marks(self):
        return sum(self.marks)

    def average_score(self):
        return sum(self.marks) / len(self.marks)


student = Student("Alice", 101, [82, 95, 78, 92])
print("Total Marks:", student.total_marks()) 
print("Average Score:", student.average_score())  
```

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

rectangle = Rectangle(12, 24)
print("Area:", rectangle.area()) 
print("Perimeter:", rectangle.perimeter()) 
```

Polymorphism:

    Create a base class Shape with an abstract method calculate_area. Derive subclasses Square and Circle from Shape that implement the calculate_area method with their specific formulas. Write a function get_total_area that takes a list of Shape objects and calculates the total area of all shapes.
    Design a function greet that takes an object as an argument. If the object has a method say_hello, call that method. Otherwise, print a generic greeting. Use this function with objects of different classes that might or might not have a say_hello method to demonstrate polymorphism.


```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass

class Square(Shape):
    def __init__(self, side_length):
        self.side_length = side_length

    def calculate_area(self):
        return self.side_length ** 2

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return math.pi * self.radius ** 2

def get_total_area(shapes):
    total_area = 0
    for shape in shapes:
        total_area += shape.calculate_area()
    return total_area


shapes = [
    Square(5),
    Circle(3),
    Square(4)
]

total_area = get_total_area(shapes)
print("Total Area:", total_area)  
```

Encapsulation:

    Create a class BankAccount with attributes like account_number (private) and balance (protected). Implement methods to deposit and withdraw funds (with appropriate checks) and a public method to get the current balance.
    Design a class Calculator with private methods for basic operations (add, subtract, multiply, divide). Provide public methods to perform these calculations on user-provided numbers.


```python
class BankAccount:
    def __init__(self, account_number, initial_balance=0):
        self.__account_number = account_number  # Private attribute
        self._balance = initial_balance  # Protected attribute

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited: ${amount}")
        else:
            print("Deposit amount must be positive")

    def withdraw(self, amount):
        if amount > 0 and amount <= self._balance:
            self._balance -= amount
            print(f"Withdrew: ${amount}")
        elif amount > self._balance:
            print("Insufficient funds")
        else:
            print("Withdrawal amount must be positive")

    def get_balance(self):
        return self._balance


account = BankAccount("422456789", 1700)
account.deposit(1200)
account.withdraw(200)
print("Current Balance:", account.get_balance())
```

```python
class Calculator:
    def __add(self, a, b):
        return a + b

    def __subtract(self, a, b):
        return a - b

    def __multiply(self, a, b):
        return a * b

    def __divide(self, a, b):
        if b != 0:
            return a / b
        else:
            return "Cannot divide by zero"

    def add(self, a, b):
        return self.__add(a, b)

    def subtract(self, a, b):
        return self.__subtract(a, b)

    def multiply(self, a, b):
        return self.__multiply(a, b)

    def divide(self, a, b):
        return self.__divide(a, b)

calculator = Calculator()
print("Add:", calculator.add(15, 5))       
print("Subtract:", calculator.subtract(15, 5))  
print("Multiply:", calculator.multiply(15, 5)) 
print("Divide:", calculator.divide(15, 5))     
print("Divide by zero:", calculator.divide(15, 0))  
```

 Shape Hierarchy and Polymorphism:

Develop a hierarchy of shape classes with an abstract base class Shape containing an abstract method calculate_area(). Create subclasses like Square, Circle, Rectangle, and Triangle that override the calculate_area() method with their specific formulas.

Challenge: Implement a function draw_shape(shape) that takes a Shape object as an argument and uses polymorphism to call the appropriate drawing method (e.g., using external libraries like Matplotlib) based on the shape type.

```python
import matplotlib.pyplot as plt
import numpy as np

class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass

    @abstractmethod
    def draw(self):
        pass

class Square(Shape):
    def __init__(self, side_length):
        self.side_length = side_length

    def calculate_area(self):
        return self.side_length ** 2

    def draw(self):
        plt.figure()
        square = plt.Rectangle((0, 0), self.side_length, self.side_length, color='blue')
        plt.gca().add_patch(square)
        plt.axis('equal')
        plt.title('Square')
        plt.show()

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return math.pi * self.radius ** 2

    def draw(self):
        plt.figure()
        circle = plt.Circle((0, 0), self.radius, color='red')
        plt.gca().add_patch(circle)
        plt.axis('equal')
        plt.title('Circle')
        plt.show()

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

    def draw(self):
        plt.figure()
        rectangle = plt.Rectangle((0, 0), self.length, self.width, color='green')
        plt.gca().add_patch(rectangle)
        plt.axis('equal')
        plt.title('Rectangle')
        plt.show()

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def calculate_area(self):
        return 0.5 * self.base * self.height

    def draw(self):
        plt.figure()
        triangle = plt.Polygon([[0, 0], [self.base, 0], [self.base/2, self.height]], color='purple')
        plt.gca().add_patch(triangle)
        plt.axis('equal')
        plt.title('Triangle')
        plt.show()

def draw_shape(shape):
    shape.draw()


square = Square(5)
circle = Circle(3)
rectangle = Rectangle(4, 6)
triangle = Triangle(4, 5)

draw_shape(square)
draw_shape(circle)
draw_shape(rectangle)
draw_shape(triangle)
```

File and Data Management System:

Design classes to manage files and data in various formats:

    FileHandler: Attributes: filename, data (list/dictionary depending on format). Methods: read_data(), write_data(), check_file_exists().
    CSVHandler (subclass of FileHandler): Methods: parse_csv(), write_csv() (specific to CSV format).
    JSONHandler (subclass of FileHandler): Methods: parse_json(), write_json() (specific to JSON format).

Challenge: Implement functionality to handle different file formats (CSV, JSON) using inheritance and polymorphism. Ensure proper error handling for invalid file formats or data structures.

```python
import csv
import json
import os

class FileHandler:
    def __init__(self, filename):
        self.filename = filename
        self.data = None

    def read_data(self):
        try:
            with open(self.filename, 'r') as file:
                self.data = file.read()
                print(f"Data read from {self.filename}:")
                print(self.data)
        except FileNotFoundError:
            print("File not found.")
            self.data = None

    def write_data(self):
        if self.data is not None:
            try:
                with open(self.filename, 'w') as file:
                    file.write(self.data)
                print(f"Data written to {self.filename}")
            except Exception as e:
                print("Error writing data:", e)
        else:
            print("No data to write.")

    def check_file_exists(self):
        return os.path.isfile(self.filename)

class CSVHandler(FileHandler):
    def __init__(self, filename):
        super().__init__(filename)

    def parse_csv(self):
        try:
            with open(self.filename, 'r') as file:
                reader = csv.reader(file)
                self.data = list(reader)
                print(f"Data parsed from CSV {self.filename}:")
                print(self.data)
        except Exception as e:
            print("Error parsing CSV:", e)

    def write_csv(self):
        if self.data is not None:
            try:
                with open(self.filename, 'w', newline='') as file:
                    writer = csv.writer(file)
                    writer.writerows(self.data)
                print(f"Data written to CSV {self.filename}")
            except Exception as e:
                print("Error writing CSV:", e)
        else:
            print("No data to write.")

class JSONHandler(FileHandler):
    def __init__(self, filename):
        super().__init__(filename)

    def parse_json(self):
        try:
            with open(self.filename, 'r') as file:
                self.data = json.load(file)
                print(f"Data parsed from JSON {self.filename}:")
                print(self.data)
        except Exception as e:
            print("Error parsing JSON:", e)

    def write_json(self):
        if self.data is not None:
            try:
                with open(self.filename, 'w') as file:
                    json.dump(self.data, file, indent=4)
                print(f"Data written to JSON {self.filename}")
            except Exception as e:
                print("Error writing JSON:", e)
        else:
            print("No data to write.")

# Create example CSV file if it doesn't exist
def create_csv_file(filename):
    if not os.path.isfile(filename):
        with open(filename, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['Name', 'Age', 'City'])
            writer.writerow(['John', '30', 'New York'])
            writer.writerow(['Alice', '25', 'Los Angeles'])
            writer.writerow(['Bob', '35', 'Chicago'])
        print(f"Example CSV file {filename} created with sample data.")

# Create example JSON file if it doesn't exist
def create_json_file(filename):
    if not os.path.isfile(filename):
        data = {
            "users": [
                {"name": "John", "age": 30, "city": "New York"},
                {"name": "Alice", "age": 25, "city": "Los Angeles"},
                {"name": "Bob", "age": 35, "city": "Chicago"}
            ]
        }
        with open(filename, 'w') as file:
            json.dump(data, file, indent=4)
        print(f"Example JSON file {filename} created with sample data.")


csv_filename = "data.csv"
json_filename = "data.json"

create_csv_file(csv_filename)
create_json_file(json_filename)

# Using FileHandler subclasses to read and write data
csv_handler = CSVHandler(csv_filename)
json_handler = JSONHandler(json_filename)

csv_handler.parse_csv()
json_handler.parse_json()

# Modify data (assuming data is already loaded)
if csv_handler.data:
    csv_handler.data.append(['New', '40', 'San Francisco'])
if json_handler.data:
    json_handler.data['users'].append({"name": "Eve", "age": 28, "city": "Boston"})

# Write modified data back to files
csv_handler.write_csv()
json_handler.write_json()
```
